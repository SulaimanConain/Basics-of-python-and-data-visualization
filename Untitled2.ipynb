```python
#stackplot or Area plot
#how to merge two plots?
import matplotlib.pyplot as plt
%matplotlib inline
y1 = cars['hp']
y2 = cars['disp']
x = range(32)
plt.stackplot(x,y1,alpha=1,color='orange')
plt.stackplot(x,y2,alpha=0.5,color='green')  
```

```python
#how to merge two plots?
import matplotlib.pyplot as plt
%matplotlib inline
y1 = cars['hp']
y2 = cars['disp']
x = range(32)
plt.plot(x,y1,linewidth=2.0,color='r')
plt.stackplot(x,y1, color='purple',alpha=0.7)
plt.plot(x,y1, linewidth=1.0,color='c')
plt.stackplot(x,y2,color='black',alpha=0.5)
```
